```python
import pandas as pd
import os
```

```python
folder_path = '/content/drive/MyDrive/MLOPS-zoomcamp/'
```

```python
parquet_files = sorted(
    [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.parquet')]
)
```

```python
parquet_files
```

```python
df_1 = pd.read_parquet(parquet_files[0])
```

```python
df_1.head()
```

```python

df_1.info()
```

```python
df_1['duration'] =df_1['tpep_dropoff_datetime'] - df_1['tpep_pickup_datetime']
df_1['duration'] =df_1['duration'].dt.total_seconds()/60
```

```python
df_1
```

```python
df_1['duration'].std()
```

```python
df  = df_1[(df_1.duration >= 1) & (df_1.duration <= 60)]
```

```python
len(df)/len(df_1)
```

```python
categorical = ['PULocationID', 'DOLocationID']
df[categorical] = df[categorical].astype(str)
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dicts = df[categorical].to_dict(orient='records')
```

```python
dv= DictVectorizer()
x_train= dv.fit_transform(train_dicts)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
y_train = df['duration'].values
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
rmse = root_mean_squared_error(y_train, lr.predict(x_train))
rmse
```

```python
# validation dataset  RMSE

df_2 = pd.read_parquet(parquet_files[1])
df_2['duration'] =df_2['tpep_dropoff_datetime'] - df_2['tpep_pickup_datetime']
df_2['duration'] =df_2['duration'].dt.total_seconds()/60
df_new  = df_2[(df_2.duration >= 1) & (df_2.duration <= 60)]
df_new[categorical] = df_new[categorical].astype(str)
val_dicts = df_new[categorical].to_dict(orient='records')
x_val= dv.transform(val_dicts)
y_val = df_new['duration'].values
rmse_val= root_mean_squared_error(y_val, lr.predict(x_val))
rmse_val

```
